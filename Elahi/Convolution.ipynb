```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convolve(x,h):
    len_x=len(x)
    len_h=len(h)
    len_y=len_x+len_h-1
    y=np.zeros(len_y)
    for i in range(len_y):
        for k in range (len_x):
            if(i-k>=0 and i-k<len_h):
                y[i]+=x[k]*h[i-k]
    return y         
```

```python
x = [1, 2, 3, 1]
h = [1, 2, 1, -1]

plt.figure(figsize=(8,7))
plt.subplot(3,1,1)
plt.stem(x)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("x(n)")

plt.subplot(3,1,2)
plt.stem(h)
plt.xlabel("Time()")
plt.ylabel("Amplitude")
plt.title("h(n)")
plt.tight_layout()
y=np.convolve(x,h)

plt.subplot(3,1,3)
plt.stem(y)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("y(n)")
plt.tight_layout()
```

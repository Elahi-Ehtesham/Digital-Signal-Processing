```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Unit impulse signal
t=np.arange(-5,5)
x=np.where(t==0,1,0)

plt.figure(figsize=(10,4))
plt.stem(t,x)
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title("Unit impulse signal")
```

```python
#Unit step signal
t=np.arange(-5,5)
x=np.where(t>=0,1,0)

plt.figure(figsize=(10,4))
plt.stem(t,x)
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title("Unit step signal")
```

```python
#Unit ramp signal
t=np.arange(-5,10)
x=np.where(t>=0,t,0)

plt.figure(figsize=(10,4))
plt.stem(t,x)
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title("Unit Ramp signal")
```

```python
#Exponential signal when 0<a<1
n=np.arange(0,20)
a=0.8
x=a**n

plt.figure(figsize=(10,4))
plt.stem(n,x)
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title("Exponential signal 0<a<1")
```

```python
#Exponential signal when a>1
n=np.arange(0,20)
a=1.1
x=a**n

plt.figure(figsize=(10,4))
plt.stem(n,x)
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title("Exponential signal a>1")
```

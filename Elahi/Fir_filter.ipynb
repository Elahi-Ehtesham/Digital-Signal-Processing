```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz

# Given
fs = 8000
fc = 1000
N = 21
M = (N - 1) // 2

# Normalized cutoff
wc = 2 * np.pi * fc / fs

# Ideal impulse response
n = np.arange(N)
hd = np.zeros(N)

for i in range(N):
    if i == M:
        hd[i] = wc / np.pi
    else:
        hd[i] = np.sin(wc * (i - M)) / (np.pi * (i - M))

# Hamming window
w = 0.54 - 0.46 * np.cos(2 * np.pi * n / (N - 1))

# Final filter
h = hd * w

# Frequency response
w_freq, H = freqz(h, worN=8000)

# Plot Impulse Response
plt.figure()
plt.stem(n, h)
plt.title("Impulse Response")
plt.xlabel("n")
plt.ylabel("h[n]")

# Plot Magnitude Response
plt.figure()
plt.plot(w_freq * fs / (2 * np.pi), abs(H))
plt.title("Magnitude Response")
plt.xlabel("Frequency (Hz)")
plt.ylabel("|H(f)|")

# Plot Phase Response
plt.figure()
plt.plot(w_freq * fs / (2 * np.pi), np.angle(H))
plt.title("Phase Response")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase")

plt.show()
```

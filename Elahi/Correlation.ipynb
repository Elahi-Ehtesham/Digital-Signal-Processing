```python
import numpy as np
import matplotlib.pyplot as plt
print("")
```

```python
def convolve(x,h):
    len_x=len(x)
    len_h=len(h)
    len_y=len_x+len_h-1
    y=np.zeros(len_y)
    for i in range (len_y):
        for k in range (len_x):
            if(i-k>=0 and i-k<len_h):
                y[i]+=x[k]*h[i-k]
    return y
```

```python
t=np.arange(0,1,0.01)
x=np.sin(5*np.pi*t)
delay=10

#delay signal without increasing length
x_delayed=np.concatenate((np.zeros(delay),x[:-delay]))
corr=convolve(x,np.flip(x_delayed))

plt.figure(figsize=(10,4))

plt.subplot(2,2,1)
plt.plot(x)
plt.title("Original signal")
plt.tight_layout()

plt.subplot(2,2,2)
plt.plot(x_delayed)
plt.title("Delayed signal")

plt.subplot(2,2,3)
plt.plot(corr)
plt.title("Cross-correlation")
plt.xlabel("lag")

lags=np.arange(-len(x)+1,len(x))
peak_index=np.argmax(corr)
estimated_delay=lags[peak_index]

print("Actual delay: ",delay)
print("Estimated delay: ",estimated_delay)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dft(xn):
    N=len(xn)
    xm=np.zeros(N,dtype=complex)
    for i in range(N):
        for k in range(N):
            xm[i]+=xn[k]*np.exp(-2j*np.pi*i*k/N)
    return xm
```

```python
def phase(xm):
    ph=np.zeros(len(xm))
    for i in range(len(xm)):
        ph[i]=np.arctan2(round(xm[i].imag),round(xm[i].real))
    return np.degrees(ph)
```

```python


fs=100
t=np.arange(0,1,1/fs)
x1n=np.sin(2*np.pi*30*t)


plt.figure(figsize=(10,9))
plt.subplot(3,2,1)
plt.stem(t,x1n)
plt.xlabel("Time(t)")
plt.ylabel("Amplitude")
plt.title("sampled signal")

x2n=np.sin(2*np.pi*70*t)

plt.subplot(3,2,2)
plt.stem(t,x2n)
plt.xlabel("Time(t)")
plt.ylabel("Amplitude")
plt.title("sampled signal")

x1m=dft(x1n)

plt.subplot(3,2,3)
plt.stem(np.abs(x1m))
plt.xlabel("frequency")
plt.ylabel("Amplitude")
plt.title("Magnitude spectrum")
plt.tight_layout()
x2m=dft(x2n)

plt.subplot(3,2,4)
plt.stem(np.abs(x2m))
plt.xlabel("frequency")
plt.ylabel("Amplitude")
plt.title("Magnitude spectrum")
plt.tight_layout()

ph1=phase(x1m)
plt.subplot(3,2,5)
plt.stem(ph1)
plt.xlabel("frequency")
plt.ylabel("Phase")
plt.title("phase spectrum")
plt.tight_layout()

```

```python
def hamming_window(N):
    w=np.zeros(N)
    for i in range(N):
        w[i]=0.54-0.46*np.cos(2*np.pi*i/(N-1))
    return w
```

```python
t=np.arange(0,1,0.001)
x=np.sin(2*np.pi*10*t)

plt.figure(figsize=(10,5))
plt.subplot(3,2,1)
plt.plot(t,x)

fs=100

n=np.arange(0,1,1/fs)
xn=np.sin(2*np.pi*10*n)
plt.subplot(3,2,2)
plt.stem(n,xn)


x1m=dft(xn)

plt.subplot(3,2,3)
plt.stem(np.abs(x1m))
plt.xlabel("frequency")
plt.ylabel("Amplitude")
plt.title("Magnitude spectrum")
plt.tight_layout()

n2=np.arange(0,0.95,1/fs)
xn2=np.sin(2*np.pi*10*n2)

xm2=dft(xn2)
plt.subplot(3,2,4)
plt.stem(np.abs(xm2))
plt.xlabel("frequency")
plt.ylabel("Amplitude")
plt.title("Magnitude spectrum")
plt.tight_layout()

w=hamming_window(len(xn2))
x2w=xn2*w[:len(xn2)]

x2wm=dft(x2w)

plt.subplot(3,2,5)
plt.stem(np.abs(x2wm))
plt.xlabel("frequency")
plt.ylabel("Amplitude")
plt.title("Magnitude spectrum")
plt.tight_layout()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz
def con(x,h):
    lenx=len(x)
    lenh=len(h)
    leny=lenx+lenh-1
    y=np.zeros(leny)
    for i in range(leny):
        for k in range(lenx):
            if i-k>=0 and i-k<lenh:
                y[i]+=x[k]*h[i-k]
    return y            
```

```python


fs=100
t=np.arange(0,1,1/fs)
f=10
xclean=np.sin(2*np.pi*10*t)

noise=0.5*np.random.randn(len(t))

xnoise=xclean+noise

N=6
h=np.ones(N)/N
y=con(xnoise,h)
#y=np.convolve(xnoise,h)

plt.figure(figsize=(10,4))
plt.subplot(2,2,1)
plt.plot(xnoise)
plt.title("Noisy Signal")

plt.subplot(2,2,2)
plt.plot(y)
plt.title("Filtered Signal")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Aliasing
t=np.arange(0,0.1,0.001)
x1=np.sin(200*np.pi*t)
x2=np.sin(300*np.pi*t)

plt.figure(figsize=(10,4))

plt.subplot(2,2,1)
plt.plot(t,x1)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("x(t)=sin(200pit)")
plt.tight_layout()

plt.subplot(2,2,2)
plt.plot(t,x2)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("x(t)=sin(300pit)")
plt.tight_layout()

#Sampling frequency
fs=75
n=np.arange(0,0.1,1/fs)

x1_n=np.sin(200*np.pi*n)
plt.subplot(2,2,3)
plt.plot(n,x1_n)
plt.xlabel("Discrete time")
plt.ylabel("Amplitude")
plt.title("Sampled x1")
plt.tight_layout()

x2_n=np.sin(300*np.pi*n)
plt.subplot(2,2,4)
plt.plot(n,x1_n)
plt.xlabel("Discrete time")
plt.ylabel("Amplitude")
plt.title("Sampled x2")
plt.tight_layout()
```

Generate a cosine signal represented as x[t]=3cos( 100 pi t).
1.sample the generated signal at 75 Hz and 200 Hz ,Plot the sampled signal

```python
#Sampling example
t=np.arange(0,0.1,0.001)
x=3* np.cos(np.pi*100*t)

plt.figure(figsize=(10,4))
plt.subplot(2,2,1)
plt.plot(t,x)
plt.xlabel("Time(s)")
plt.ylabel("amplitude")
plt.title("Graph of X")
plt.tight_layout()
# y=5* np.sin(2*np.pi*30*t)

# plt.subplot(2,2,2)
# plt.plot(t,y)
# plt.xlabel("Time(s)")
# plt.ylabel("Amplitude")
# plt.title("Graph of Y")

fs=75
n=np.arange(0,0.1,1/fs)
x_n=3* np.cos(100*np.pi*n)

plt.subplot(2,2,2)
plt.stem(n,x_n)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Sampled signal at 75 hz")
plt.tight_layout()

fs2=200
n2=np.arange(0,0.1,1/fs2)
x2=3*np.cos(100*np.pi*n2)

plt.subplot(2,2,3)
plt.stem(n2,x2)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Sampled at 200Hz")
plt.tight_layout()

```
